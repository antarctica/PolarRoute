```python
import pandas as pd
import numpy as np
from refactor.CellBox import CellBox
from matplotlib.patches import Polygon
import matplotlib.pylab as plt
from refactor.CellGrid import CellGrid
import matplotlib.pyplot as pltSave

#read in IcePoints
icePoints = pd.read_json('resources/icepoints.json')
icePoints = pd.DataFrame.from_records(icePoints.icepoints)

#read in CurrentPoints
currentPoints = pd.read_json('resources/currentPoints.json')
currentPoints = pd.DataFrame.from_records(currentPoints.currentPoints)
```

```python
#Initialize cell grid
longMin = 260
longMax = 350
latMin = -75
latMax = -50

cellWidth = 5
cellHeight = 2.5

cellGrid = CellGrid(longMin, longMax, latMin, latMax, cellWidth, cellHeight)

#add icepoints and currentpoints to cell grid
cellGrid.addIcePoints(icePoints)
cellGrid.addCurrentPoints(currentPoints)
```

```python
cellGrid.recursiveSplit(4)
cellGrid.plot()
```

```python
longMin = 330
longMax = 350
latMin = -75
latMax = -70

cellWidth = 1
cellHeight = 0.5

cellGridSection = CellGrid(longMin, longMax, latMin, latMax, cellWidth, cellHeight)
cellGridSection.addIcePoints(icePoints)
cellGridSection.addCurrentPoints(currentPoints)
```
