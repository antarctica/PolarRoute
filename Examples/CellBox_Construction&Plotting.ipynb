```python
import pandas as pd
from RoutePlanner.CellGrid import CellGrid

#read in IcePoints
icePoints = pd.read_json('../resources/Jan2022_InitialBoxCreation/icepoints2019-02-08.json')
icePoints = pd.DataFrame.from_records(icePoints.icepoints)

#read in CurrentPoints
currentPoints = pd.read_json('../resources/Jan2022_InitialBoxCreation/currentPoints.json')
currentPoints = pd.DataFrame.from_records(currentPoints.currentPoints)
```

```python
longMin = 260
longMax = 350
latMin = -75
latMax = -50

cellWidth = 5
cellHeight = 2.5

cellGrid = CellGrid(longMin, longMax, latMin, latMax, cellWidth, cellHeight)
cellGrid.addIcePoints(icePoints)
cellGrid.addCurrentPoints(currentPoints)
```

```python
cellGrid.plot()
```

```python
cellGrid.recursiveSplit(3)
cellGrid.plot()
```

```python
selectedCell = cellGrid.getCellBox(-61,296)
cellGrid.highlightCell(selectedCell)
```

```python
neightbours = cellGrid.getNeightbours(selectedCell)
cellGrid.highlightCells(neightbours)
```
