```python
def load_graph(path):
    f = open(path, "r")

    dumpGraph = {}

    for line in f:
        parts = line.split(",")
        node = parts[0] + "," + parts[1]
        neighbours = ""
        for x in range (2, len(parts)):
            neighbours += parts[x] + ","
        neighbours = neighbours[:-2]  # remove last comma and newline from neighbours
    
        dumpGraph[node] = neighbours
    f.close()
    return dumpGraph
```

```python
def compare_graphs(java, python):
    diff = {}

    for node in java.keys():
        if node in python.keys():
            if java[node] == python[node]:
                diff[node] = "Match"
            else:
                diff[node] = "Diff"
        else:
            diff[node] = "Not Found"
    return diff
```

```python
javaDumpGraph = load_graph("JavaGraphDump.csv")
pythonDumpGraph = load_graph("PythonGraphDump.csv")
```

```python
javaDumpGraph
```

```python
comparrison = compare_graphs(javaDumpGraph, pythonDumpGraph)

matchCount = 0
diffCount = 0
notFoundCount = 0 

diff = []

for node in comparrison:
    if comparrison[node] == "Match":
        matchCount += 1
    if comparrison[node] == "Diff":
        diffCount += 1
        error = {}
        error["node"] = node
        error["Java"] = javaDumpGraph[node]
        error["Python"] = pythonDumpGraph[node]
        diff.append(error)
    if comparrison[node] == "Not Found":
        notFoundCount += 1
        
print("Matches = " + str(matchCount))
print("Differences = " + str(diffCount))
print("Not Found = " + str(notFoundCount))
```

```python
for node in comparrison:
    if comparrison[node] == "Not Found":
        print(node)
```

```python
for x in diff:
    print("Node   : " + str(x["node"]))
    print("Java   : " + str(x["Java"]))
    print("Python : " + str(x["Python"]))
    print("")
```
