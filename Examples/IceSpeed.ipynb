```python
import matplotlib.pyplot as plt
from numpy import polyval, polyfit
```

```python
# Data from Zhang et al., Ocean Engineering 186 (2019) 106071

top_speed = 14

speeds = [14, 9.5, 9.0, 5.5, 5.0, 0.0]
cons = [0, 0.1, 0.3, 0.5, 0.7, 0.8]

norm_speeds = [s/top_speed for s in speeds]

plt.scatter(cons, norm_speeds)
plt.xlabel("Ice Concentration")
plt.ylabel("Normalised speed")
```
